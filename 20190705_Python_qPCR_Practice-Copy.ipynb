```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
#Process the raw data output and fill NA values with 0.
data = pd.read_excel ("data/2019-02-27_161601 AtWAKL8 different version expressions.xls",
                     sheet_name='Results').fillna(0)
data.head(3)
```

```python
# iloc[row slicing, column slicing]
data = data.iloc[46:191, 3:15]

#Create a new header row for the columns
data = data.rename(columns=data.iloc[0]).drop(data.index[0])

#Omit all columns except for "Sample Name", "Target Name" and "CT"
data = data[['Sample Name', 'Target Name', 'CT']]
data.reset_index(drop=True)
data.head(3)
```
